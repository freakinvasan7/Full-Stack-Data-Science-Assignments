#Q3 What is the most significant distinction between a dictionary and a list?
Ans. Key value pairing. In a list all the elements are stored as values while in a dictionary all elements come in key value pairs, where every value has an associated key mapped to it.

#Q4 What happens if you try to access spam 'foo' if spam is {'bar': 100}?
Ans. We get a key error as there is no key named 'foo' in the dictionary spam.

#Q5 If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and
'cat' in spam.keys()?
Ans. The output in both scenarios will be the same however both methods check the same in a different way. 'cat' in spam checks directly if the key is present in the dictionary or not while 'cat' in spam.keys() check the list of keys generated by the command for 'cat' in it.

#Q6 If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and
'cat' in spam.values()?
Ans. Both are completely different. 'cat' in spam will check directly if there is key which has value 'cat' however 'cat' in spam.values() checks in the list of values of the dictionary if 'cat' is present or not.
